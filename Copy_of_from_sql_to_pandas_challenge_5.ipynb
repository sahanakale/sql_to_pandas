## 5. Select the name and address of all stores located in an Avenue or in a Street (address ended in "St.")

````python
df5 = stores[(stores["stor_address"].str.contains("Ave.")) | (stores["stor_address"].str.contains("St."))]
df5[['stor_name', 'stor_address']]
````

### Hint:

In SQL the syntax is:

```sql
SELECT
	stor_name,
	stor_address
FROM
	stores
WHERE 
	stor_address LIKE '%St.' 
    OR stor_address LIKE '%Ave.';
```

## 7. Select the first and last names of all male employees whose name starts with "P".

````python
df5 = employee[((employee["fname"].str.contains("^P")) | (employee["lname"].str.contains("^P"))) & (employee["emp_id"].str.contains("M$"))]
df5[['emp_id', 'fname', 'lname']]
````

### Hint:

In SQL the syntax is:

```sql
SELECT
    emp_id,
    fname,
    lname
FROM
    employee
WHERE
    (fname LIKE 'P%' 
    OR lname LIKE 'P%') 
    AND emp_id LIKE '%M';
```
